```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import  StandardScaler
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
```

```python

data=pd.read_csv(r'C:\Users\cheth\Assignment\Clustering\crime_data.csv')
print (data.head())
```

```python
data.head()
```

```python
data1=data.rename(columns={"Unnamed: 0":"City"})
```

```python
print (data1.info())
```

```python
data1.drop(['City'],axis=1,inplace=True)
```

```python
data1
```

```python
#rray=df.values
array=data1.values
```

```python
array
```

```python
stscaler=StandardScaler().fit(array)
```

```python
stscaler
```

```python
x=stscaler.transform(array)
```

```python
dbscan=DBSCAN(eps=1.2,min_samples=4)
dbscan.fit(x)
```

```python
dbscan.labels_
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['clusters'])
```

```python
df1=pd.concat([data1,cl],axis=1)
```

```python
#Indexes of core data points
dbscan.core_sample_indices_
```

```python
df1
```

```python
df1.groupby('clusters').agg(['count']).reset_index()
```
